```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

chdf = pd.read_csv('./Desktop/Community_Health_Survey.csv')
print(chdf.columns)
```

```python
q = chdf.groupby(['Question']).groups.keys()
y = chdf.groupby(['Year']).groups.keys()
print(q)
```

```python
chdf.head(20)
```

```python
final = chdf.drop_duplicates()
```

```python
Obesity = []
BingeDrinking = []
Colon = []
Drinks = []
FluShot = []
NPD = []
HS = []
SS = []
MedicalAttn = []
HIC = []

i = {2010, 2011, 2012, 2013, 2014, 2015}
for il in i:
    for index, row in final.iterrows():
        if((row['Question'] == 'Obesity') and (row['Year'] == il)):
            Obesity.append(row['Prevalence'])
        if((row['Question'] == 'Binge Drinking') and (row['Year'] == il)):
            BingeDrinking.append(row['Prevalence'])
        if((row['Question'] == 'Colon cancer screening, adults age 50+ (colonoscopy)') and (row['Year'] == il)):
            Colon.append(row['Prevalence'])
        if((row['Question'] == 'Did not get needed medical care') and (row['Year'] == il)):
            MedicalAttn.append(row['Prevalence'])
        if((row['Question'] == 'Drinks 1 or more sugar-sweetened beverages per day') and (row['Year'] == il)):
            Drinks.append(row['Prevalence'])
        if((row['Question'] == 'Flu shot in last 12 months, adults ages 65+ (not age-adjusted)') and (row['Year'] == il)):
            FluShot.append(row['Prevalence'])
        if((row['Question'] == 'No Personal Doctor') and (row['Year'] == il)):
            NPD.append(row['Prevalence'])
        if((row['Question'] == 'Self-reported Health Status (excellent/very good/good)') and (row['Year'] == il)):
            HS.append(row['Prevalence'])
        if((row['Question'] == 'Smoking Status (current smokers)') and (row['Year'] == il)):
            SS.append(row['Prevalence'])
        if((row['Question'] == 'Health Insurance Coverage') and (row['Year'] == il)):
            HIC.append(row['Prevalence'])   
                            
print(Obesity) 
print(BingeDrinking)
print(Colon)
print(Drinks)
print(FluShot)
print(NPD)
print(HS)
print(SS)
print(MedicalAttn)
```
